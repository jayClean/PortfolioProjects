```python
#Importing libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # adjusts the configration of the plot we make

# Reading csv data
df = pd.read_csv(r"C:\Users\Rohit Mohanty\Downloads\movies.csv\movies.csv")
```

```python
#Glimpse at the data
df.head(10)
```

```python
#finding if there's any missing data
for col in df.columns:
    pct = np.mean(df[col].isnull())
    print('{}-{}'.format(col,pct))
```

```python
df.head()
```

```python
#dropping the null values rows
df.dropna(inplace=True)
```

```python
#types of df
df.dtypes
```

```python
#changing the type of gross,budget, votes
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')
```

```python
#coreecting year values

df['year_corrected'] = df['released'].astype(str)
```

```python
df['year_corrected'] = df['year_corrected'].str.split().str[2]
```

```python
# Scatter plot of budget vs gross

plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget Vs Gross')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()
```

```python
df = df.sort_values(by = ['gross'], inplace = False, ascending= False)
```

```python
df.head()
```

```python
#plotting budget vs gross using seaborn
sns.regplot(x = 'budget', y ='gross', data = df, scatter_kws = {'color':'red'}, line_kws = {'color':'blue'})
```

```python
#correlation--methods = pearson, kendall, spearman

df.corr(method='pearson')
```

```python
#high correaltion b/w budget and gross
#visulaization

correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized = df
for col in df_numerized:
    if (df_numerized[col].dtype == 'object'):
        df_numerized[col] = df_numerized[col].astype('category')
        df_numerized[col] = df_numerized[col].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
corr_mat = df_numerized.corr()
corr_pairs = corr_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>=0.5]
high_corr = high_corr[(high_corr)!=1]
high_corr
```
